```python
#### Vectorization
# Perform ops on entire arrays or array elements simultaneously which is significantly faster than using explicit loops
import numpy as np

def myfunc(a, b):
    "Return a-b if a>b, otherwise return a+b"
    if a > b:
        return a - b
    else:
        return a + b

vfunc = np.vectorize(myfunc)
vfunc([1, 2, 3, 4], 2)
```

```python
def find_square(x):
    if x < 0:
        return 0
    else:
        return x ** 2

vectorized_sq = np.vectorize(find_square)
array1 = np.array([-1, 0, 2, 3, 4])
result = vectorized_sq(array1)
print(result)  # [0 0 4 9 16]
```

```python
### Broadcasting

# Rule 1: Pad Leading Dimensions
  # Shorter arrays pad with ones on the left to match the longer array's rank.
  #Example: Shape (3,) pads to (1,3) when added to (4,3).

# Rule 2: Stretch Size-1 Dimensions
   # In each dimension, if sizes differ, the size-1 dimension stretches to match the other.
   # Example: (5,1) + (1,7) → both stretch to (5,7).

# Rule 3: Exact Match or Fail
   # Dimensions must be equal or 1; otherwise, raise ValueError.
   # Example: (3,3) + (2,3) fails (3≠2 and neither is 1).
```

```python
#### **Case 1:** If dimension in both matrix is equal, element-wise addition will be done.
#### **Case 2:** Right array should be of 1-D and number of columns should be same of both the arrays and it will automatically do n-tile.
#### **Case 3:** If the left array is column matrix (must have only 1 column) and right array is row matrix, then it will do the n-tile such that element wise addition is possible.
```

```python
# Scalar Broacast
import numpy as np
a = np.array([1, 2, 3])
result = a + 10  # Scalar 10 broadcasts to [10,10,10]
print(result)  # [11 12 13]
```

```python
#A 1D array broadcasts across rows of a 2D array if trailing dims match.
import numpy as np
a = np.array([[1,2,3], [4,5,6]])  # shape (2,3)
b = np.array([10,20,30])           # shape (3,)
result = a + b
print(result)
# [[11 22 33]
#  [14 25 36]]
```

```python
# Column-wise Expansion

# A column vector (shape with 1 in column dim) stretches across columns.'
import numpy as np
a = np.array([[1,2,3], [4,5,6]])  # (2,3)
b = np.array([[10],[20]])         # (2,1)
result = a + b
print(result)
# [[11 12 13]
#  [24 25 26]]
```

```python
# Failure Case
import numpy as np
a = np.array([1,2,3])  # (3,)
b = np.array([[1,2],[3,4]])  # (2,2)
# result = a + b  # Errors: axes don't align (trailing dims 3 vs 2)
result = a + b
```

```python
# Shallow Copy -> Reshape, Slicing, view
# Deep Copy -> Arithmetic ops, masking, copy
# Exception in copy() when it behaves as shallow copy when dtype='object'. Import copy and use copy.deepcopy()

arr = np.array([1, 'm', [1,2,3]], dtype = 'object')
print("arr", arr)

copy_arr = arr.copy()
print("copy arr", copy_arr)


copy_arr[2][0] = 999

print("copy arr", copy_arr)

print("arr", arr)

import copy
copy = copy.deepcopy(arr)
copy

print("Using copy.deepcopy arr", copy)
```
